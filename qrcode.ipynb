# Ví dụ về rủi ro của QR code

```
import qrcode
qr = qrcode.QRCode(
    version=1,
    error_correction=qrcode.constants.ERROR_CORRECT_L,
    box_size=10,
    border=4,
)
qr.add_data('www.iphone12.ttt.yeah.com')
qr.make(fit=True)

img = qr.make_image(fill_color="black", back_color="white")
```

```
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
imgplot = plt.imshow(img)
plt.show()
```

```
qr2 = qrcode.QRCode(
    version=1,
    error_correction=qrcode.constants.ERROR_CORRECT_L,
    box_size=10,
    border=4,
)
qr2.add_data('www.downloadthismalware.com')
qr2.make(fit=True)

img2 = qr2.make_image(fill_color="black", back_color="white")
```
